```python
import pandas as pd

results_path = "/Users/adairkovac/dev/turbulence-CFOG-7710/calculations/calc_from_intermediate/2/results.csv"

def load(path):
    return pd.read_csv(path, sep="\t", parse_dates=["time"])

results = load(results_path)
```

```python
def minus_min(df, column):
    minimum = df[column].min()
    return df[df[column] != minimum]

minus_two_mins = minus_min(minus_min(results, "L"), "L")

def row_with_minimum(df, column):
    minimum = df[column].min()
    return df[df[column] == minimum]

row_with_minimum(minus_two_mins, "L")
```
